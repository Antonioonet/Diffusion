```python
@inplace
def transformi(df): df[x] = [TF.resize(TF.to_tensor(o), (32,32)) for o in df[x]]

td = dsd.with_transform(transformi)
```

```python
bs = 256
dl = DataLoaders.from_dd(td,bs, num_workers=8)
dt = dl.train
xb,yb = next((iter(dt)))
xb[0].shape,yb
```

```python
def noisefy(x,t_steps=1000,bmin=0.0001,bmax=0.02):
  device = x.device
  n = x.shape[0]
  eps = torch.randn(x.shape,device=device)
  t = torch.randint(0,t_steps,(n,), dtype=torch.long).to(device)
  beta = torch.linspace(bmin,bmax,t_steps).to(device)
  alfa = 1 - beta
  alfa_bar = torch.cumprod(alfa,dim=0)[t].reshape(-1,1,1,1).to(device)

  x_t = alfa_bar.sqrt()*x + (1-alfa_bar).sqrt()*eps
  return ((x_t,t),eps)
```

```python
 def plot_dif(idx):
    fig, axes = plt.subplots(1, 2, figsize=xb[0][0].shape)
    org,noisy = xb[idx][0],ex[0][0][idx][0]
    org,noisy = map(to_cpu,[org,noisy])
    # Plot first image
    axes[0].imshow(noisy,cmap='gray')
    axes[0].set_title('Image 1')
    axes[0].axis('off')  # hides the axis

    # Plot second image
    axes[1].imshow(org,cmap='gray')
    axes[1].set_title('Image 2')
    axes[1].axis('off')
    print(((org-noisy)**2).sum())
    print
    plt.tight_layout()
    plt.show()

ex = noisefy(xb.to(device))
idx = torch.argmin((ex[0][1]),dim=0).to(device)
plot_dif(idx)
```

```python
# um modelo que recebe  imagem poluida e o "estágio" da poluição e retorna um output
# da dimensão da imagem original (nesse caso, vai ser a predição da poluição)
model = UNet2DModel(in_channels=1,out_channels=1,block_out_channels=(32,64,128,128)).to(device)
```

```python
epochs = 5
lr = 4e-3
loss_fuc = nn.MSELoss()
opt = optim.Adam(model.parameters(),lr=lr)
tmax = len(dl.train) *epochs
time_steps = 1000 # time steps for the diffusion process
sched = lr_scheduler.OneCycleLR(opt,max_lr=lr,total_steps=tmax)
```

```python
# treinando para aprender onde está a poluição
for epoch in range(epochs):

  i = 0
  for xb,yb in dt:
    xb = xb.to(device)
    noisy_inp = noisefy(xb,t_steps=time_steps)
    #import code; code.interact(local=locals())
    preds = model(*noisy_inp[0]).sample
    loss = loss_fuc(preds,noisy_inp[1])
    loss.backward()
    opt.step()
    sched.step()
    i += 1
    if i %10 == 0: print(f'step: {i}, train loss: {loss}')

  with torch.no_grad():
    tot_loss = 0.
    n = 0
    for xb, yb in dl.valid:
      xb = xb.to(device)
      noisy_inp = noisefy(xb,t_steps=time_steps)
      preds = model(*noisy_inp[0]).sample
      tot_loss += loss_fuc(preds,noisy_inp[1])
      n += len(xb)
    print(f'loss :{tot_loss/n}')

```
